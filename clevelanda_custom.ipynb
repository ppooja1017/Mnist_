```python
#data collection/Preprocessing
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
heart = pd.read_csv('clevelanda.csv')
```

```python
heart.head()
```

```python
#data cleaning
heart.info() # 11& 12 are objects
```

```python
heart = heart.replace('?', np.nan)
```

```python
heart = heart.dropna()
```

```python
#split data into input and output
# since there are objects/ special characters in the data, convert into numerical values
heart['class'].value_counts() # 0 = no heart disease, 1 = heart disease 
#denotes classified 
```

```python
x = heart.iloc[:, 0:13]
```

```python
x.head()
```

```python
y = heart.iloc[:, 13:]
y.head()
```

```python
#split data into train and test
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```
